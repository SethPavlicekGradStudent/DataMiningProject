```python
smallSlice = originalDF.iloc[0:10]

# Unique heroes and items
heroes = originalDF["hero_id"].unique()
items = pd.unique(originalDF[["item_0", "item_1", "item_2", "item_3", "item_4", "item_5"]].values.ravel())

# Initialize one-hot encoding dictionary
hero_columns = {f"hero_{hero}": 0 for hero in heroes}
item_columns = {f"item_{item}": 0 for item in items}

# Create an empty DataFrame to hold results
oneHotEncodedDF = pd.DataFrame(columns=list(hero_columns.keys()) + list(item_columns.keys()))

# Perform one-hot encoding
encoded_rows = []  # Use a list to collect rows for concatenation
for _, row in originalDF.iterrows():
    # Initialize a row with all zeros
    encoded_row = {**hero_columns, **item_columns}
    
    # Set the hero column to 1
    encoded_row[f"hero_{row['hero_id']}"] = 1
    
    # Set item columns to 1
    for item_col in ["item_0", "item_1", "item_2", "item_3", "item_4", "item_5"]:
        encoded_row[f"item_{row[item_col]}"] = 1
    
    # Collect the row
    encoded_rows.append(encoded_row)

# Concatenate all rows into a DataFrame
oneHotEncodedDF = pd.concat([oneHotEncodedDF, pd.DataFrame(encoded_rows)], ignore_index=True)

oneHotEncodedDF
```
